we are gonna build the funciton that is gonna take list and it will create a neighbour pair and count how often they occur 

```python
def get_pairwise_counts(ids: list[int]) -> dict[[tuple[int, int], int]]:
    counts = {}
    for pair in zip(ids, ids[1:]):
        counts[pair] = counts.get(pair, 0) + 1
    return counts
```

```python
stats = get_pairwise_counts(tokens)
print(stats)
```

sorting the dict based on the count 

```python
sortedx = sorted(stats.items(), key=lambda x: x[1], reverse=True)
print(sortedx[:10])
sortedx = {K:V for K,V in sortedx}
print(sortedx)
```

```python
def merge_most_occuring_pair(ids: list[int], pair: tuple[int,int],idx:int) -> list[int]:
    new_id = []

    i = 0
    while i < len(ids):
        if ids[i]== pair[0] and i+1 < len(ids) and ids[i+1]== pair[1]:
            new_id.append(idx)
            i += 2
        else:
            new_id.append(ids[i])
            i += 1
    return new_id

print(merge_most_occuring_pair([5, 6, 6, 7, 9, 1], (6, 7), 99))


```

```python
top_pair = list(sortedx.keys())[0]
print(f"Top pair to merge: {top_pair}")
print(f"Occurrences: {sortedx[top_pair]}")
tokens_updated = merge_most_occuring_pair(tokens, top_pair, 256)
print(tokens)
print(f"length of tokens: {len(tokens)}")
print(f"length of tokens_updated: {len(tokens_updated)}")
print(tokens_updated)
```

```python
vocab_size = 275
num_merges = vocab_size - 256

ids = list(tokens)
meregs = {}

for i in range(num_merges): 
    
     repalce_pair = list(sortedx.keys())[i]
     ids = merge_most_occuring_pair(ids, repalce_pair, 256 + i)
     print("merging pair:", repalce_pair + (f" into {256 + i}",))
     meregs[repalce_pair] = 256 + i
   
```

```python
def encode(text: str):
    tokens = list(text.encode("utf-8"))
    while len(tokens) >= 2:
        l = []
        i = 0
        while i < len(tokens):
            if i + 1 < len(tokens) and (tokens[i], tokens[i + 1]) in meregs:
                l.append(meregs[(tokens[i], tokens[i + 1])])
                i += 2
            else:
                l.append(tokens[i])
                i += 1
        tokens = l
    return tokens
```
